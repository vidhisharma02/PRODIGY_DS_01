```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Abhinav\Desktop\world population org.csv")
df
```

```python
sns.countplot(x='Region', data=df)
plt.title('Distribution of Region and IncomeGroup')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(x='Region',hue='IncomeGroup', data=df,bins=20)
plt.title('Distribution of Region and IncomeGroup')
plt.xlabel('Region')
plt.ylabel('IncomeGroup')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```
